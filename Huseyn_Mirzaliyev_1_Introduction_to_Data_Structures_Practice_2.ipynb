# Numbers: int, float, complex

```python
x = 10
y = -5
```

```python
### 1. x və y-nin cəmini tapın

# Həlli

print('x+y=',x+y)
```

```python
### 2. Nəticənin növü nədir? (a və b cəmin )

# Həlli

xy=x+y
print('x+y cəminin tipi:', type(xy))
```

# Dictionaries

```python
product_dict = {
    "Laptop": ["Dell", 999.99],
    "Smartphone": ["Apple", 799.99],
    "Headphones": ["Sony", 129.95],
    "Tablet": ["Samsung", 349.99]
}
```

```python
# 1.product_dict lüğətinin açarlarını (keys) və dəyərlərini (values) çap edin.

# Həlli

for key, values in product_dict.items():
  print(f'Key: {key}, Value: {values}')
```

```python
# 2. "Smartphone": ["Apple", 799.99] yazısını "Smartphone": ["Samsung", 500] olaraq dəyişdirin.

# Həlli

product_dict['Smartphone']= ["Samsung", 500]
print('Dəyişdirilmiş siyahı:',product_dict)
```

```python
# 3. "Smartphone" açarının dəyərini çap edin.
print('Smartphone açarının dəyəri:', product_dict.get('Smartphone'))
```

# Tuple

```python
my_tuple = ("Auto", "Fahrrad", "Motorrad")
```

```python
# 1. Python-da my_tuple tuplenin ilk və son elementini çap edin.

# Həlli

print('my_tuple tuplenin ilk elementi:', my_tuple[0])
print('my_tuple tuplenin son elementi:', my_tuple[-1])
```

```python
# 2 Python-da my_tuple  tuplenin ilk elementini silin.

# Həlli

to_list=list(my_tuple)
to_list.pop(0)
to_tuple=tuple(to_list)

print ('Ilk elementi silinmiş Tuple:', to_tuple, 'və onun tipi:', type(to_tuple))
```

# Set

```python
# 1 "Auto", "Fahrrad" və "Motorrad" dəyərləri ilə bir Python seti yaradın.

# Həlli


my_set={"Auto", "Fahrrad", "Motorrad"}
print('Yaradılmış set:', my_set, 'və onun tipi:', type(my_set))
```

```python
# 2 Artıq yaradılmış setə "Auto" elementini əlavə edin.

# Həlli

my_set.add('Auto')
print('"Auto" elementi əlavə edilmiş set:', my_set)
```

```python
# 3 my_set-dəki elementlərin sayını hesablayın və my_set-i çap edin.

# Həlli


print('Set:', my_set, 'və setin uzunluğu', len(my_set), 'elementdən ibarətdir')
```
