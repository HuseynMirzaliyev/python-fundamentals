```python
# 4. Aşağıda verilmiş sətiri böyük hərflərə  və kiçik hərflərə  çevirin
ad = "Hessen"

# həlli

upper_name=ad.upper()
lower_name=ad.lower()
print('UPPER Name:',upper_name)
print('LOWER Name:',lower_name)
```

```python
# 6. Aşağıda verilmiş iki sətiri (string) birləşdirin  və nəticəni çap edin
soz1 = "Python "
soz2 = "öyrənirik"

# həlli
# 1-ci üsul
yeni_soz=soz1+soz2
print(yeni_soz)

# 2-ci üsul
print('{}{}'.format(soz1,soz2))
```

```python
# 7. Aşağıda verilmiş sətirdə (string) "o" hərfini "a" ilə əvəz edin
cumle = "Hello World"

# həlli

replaced_cumle=cumle.replace('o','a')

print('Cümlə:',cumle)
print('Əvəzlənmiş cümlə:',replaced_cumle)
```

```python
# 8. Aşağıda verilmiş sətiri (string) boşluqlara görə parçalayın (bolun) və nəticəni siyahı (list) kimi saxlayın
cumle2 = "Bu bir testdir"

# həlli

seperated=cumle2.split()
print('Parçalanmış cümlə:', seperated)
print('Tipi:',type(seperated))
```

```python
# 9. Aşağıda verilmiş siyahıda (list) 2-ci elementi dəyişdirin və siyahını çap edin (print)
meyveler = ["alma", "banan", "gilas"]

# həlli
evvelki_siyahi=meyveler.copy()
meyveler[1]='kivi'
print('Əvvəlki siyahı:',evvelki_siyahi)
print('Yenilənmiş siyahı:', meyveler)

```

```python
# 10. Aşağıda verilmiş siyahıya (list) yeni element əlavə edin  və siyahının uzunluğunu hesablayın
rengler = ["qırmızı", "yaşıl"]

# həlli

rengler.append('mavi')
print('Rəng əlavə edilmiş list:',rengler)
```

```python
# 11. Aşağıda verilmiş siyahıdan (list) 1-ci elementi silin və siyahını sıralayın
reqemler = [5, 3, 8, 1]

# həlli

# reqemler.remove(5) - Bu üsuldan istifadə etmədim, çünki silinən rəqəmi bu üsulla göstərə bilməyirdim.
silinen=reqemler.pop(0)
reqemler.sort()
print('Silinmiş və siyahılı rəqəmlər:',reqemler)
print('Silinmiş rəqəm:',silinen)
```

```python
# 12. Aşağıda verilmiş iki siyahını (list) birləşdirin  və nəticəni çap edin
siyah1 = [1, 2]
siyah2 = [3, 4]

# həlli
yeni_siyahi=siyah1+siyah2
print('Birləşdirilmiş_siyahı:',yeni_siyahi)
print('Tipi:',type(yeni_siyahi))
```

```python
# 13. Aşağıda verilmiş dictionary-e  "ad" açarının dəyərini dəyişdirin və çap edin
telebe = {"ad": "Ali", "yas": 20}

# həlli

telebe.update({"ad": "Huseyn", "yas": 27})
print('Yenilənmiş siyahı',telebe)
```

```python
# 14. Aşağıda verilmiş lüğətə (dictionary) yeni key -value cütlüyü əlavə edin və açarları çap edin
kitablar = {"Xəmsə": "Nizami Gəncəvi"}

# həlli
kitablar['Danabaş kəndinin əhvalatları']='Cəlil Məmmədquluzadə'
for key in kitablar:
    print(f'Key: {key}')
```

```python

# 15. Aşağıda verilmiş lüğətdən (dictionary) "yas" açarını silin  və nəticəni çap edin
insan = {"ad": "Vəli", "yas": 25}

# həlli

insan.pop('yas')
print('Silinmiş lüğət:', insan)
```

```python

# 16. Aşağıda verilmiş lüğətin (dictionary) dəyərlərini toplayın  və nəticəni çap edin
qiymetler = {"alma": 5, "banan": 3}

# həlli
total_values=0
for value in qiymetler.values():
  total_values += value
print('Toplam dəyər:', total_values)
```

```python

# 17. Aşağıda verilmiş kortecdə (tuple) 2-ci elementi çıxarın və uzunluğunu hesablayın
kortec = ("qırmızı", "mavi", "yaşıl")

# həlli

print('Tuple uzunluğu:', len(kortec))

#Bir başa tuple-dan element çıxarmaq və ya daxil etmək olmur, ancaq listə çevirib, etmək olar
to_list=list(kortec)
to_list.pop(1)
to_tuple=tuple(to_list)
print("2-ci element çıxarılmış Tuple:", to_tuple)
print('Yeni Tuple uzunluğu:', len(to_tuple))

```

```python

# 19. Aşağıda verilmiş lüğətdə (dictionary) get metodu ilə dəyər çıxarın və çap edin
get_info = {"şəhər": "Bakı"}

# həlli

value=get_info.get('şəhər')
print('Çıxarılmış dəyər:', value)
```

```python
# 20. Aşağıda verilmiş iki siyahını (list) zip ilə birləşdirin və çap edin (print)
zip1 = [1, 2]
zip2 = ["a", "b"]

# həlli
zipped=list(zip(zip1,zip2))
print('Zip olunmuş siyahı:', zipped)

print( )

# for ilədə edə bilərik
for sira, elifba in zip(zip1,zip2):
  print(f'{elifba} hərfi əlifabada {sira}-inci herfdir')
```
